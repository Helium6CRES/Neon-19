```python
fig,ax = plt.subplots(1,2,figsize=(10,5))

ax[0].hist(simulated_data,bins=100);
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Counts')

ax[1].hist(simulated_data,bins=100);
ax[1].hist(background,bins=100)
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Counts')
ax[1].set_yscale('log')
```

```python
def fit_function(x,N,l):
    return (N*np.exp((-x*l))+5)

def final_analysis(sets,binsize,rc):
    
    simulated_data_runs = simulate_data(sets,rc)
    
    bins = np.linspace(0, 400, binsize)
    
    final_fits = []
    final_cov = []
    for i in np.arange(0,sets):
        data, data_bins = np.histogram(simulated_data_runs[i], bins=bins)
        binscenters = np.array([0.5 * (bins[i] + bins[i+1]) for i in range(len(bins)-1)])
        popt, pcov = curve_fit(fit_function, xdata=binscenters, ydata=data, p0=[data[0],rc])
        final_fits.extend([popt[1]])
        final_cov.extend([pcov])
    return np.asarray(final_fits),np.asarray(final_cov)

bins = np.linspace(0, 400, 100)
data_entries_1, bins_1 = np.histogram(simulated_data, bins=bins)

binscenters = np.array([0.5 * (bins[i] + bins[i+1]) for i in range(len(bins)-1)])


popt, pcov = curve_fit(fit_function, xdata=binscenters, ydata=data_entries_1, p0=[data_entries_1[0],rconst])
print(popt)


xspace = np.linspace(0, 400, 100000)

# Plot the histogram and the fitted function.
fig,ax = plt.subplots(1,2,figsize=(10,5))

ax[0].bar(binscenters, data_entries_1, width=bins[1] - bins[0], color='navy', label=r'Histogram entries')
ax[0].plot(xspace, fit_function(xspace, *popt), color='darkorange', linewidth=2.5, label=r'Fitted function')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Counts')

ax[1].bar(binscenters, data_entries_1, width=bins[1] - bins[0], color='navy', label=r'Histogram entries')
ax[1].plot(xspace, fit_function(xspace, *popt), color='darkorange', linewidth=2.5, label=r'Fitted function')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Counts')
ax[1].set_yscale('log')
```
